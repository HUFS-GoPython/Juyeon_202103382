## cp.2 Data : Types, Values, Variables and Names

## 변수의 타입

- boolean 블리언 : True, False (bool)
- integer 정수 : 1, 2, 3, 4, ... (int)
- floating point number 부동소수점 : 1.0, 2.0, 0.45 (float)
- string 문자열 : "apple", 'apple' (str)

-주의할 점
 1. 언더바로 시작하는 변수명
     ex) _name
 2. 언더바가 2개 있는 변수명
     ex) _name_
 3. 대문자로 시작하는 변수명
     ex) Car(보통 Class를 지칭할 때 사용)
 4. 전체가 대문자인 변수명   
     ex) CAR

숫자로 시작하는 단어와 예약어는 변수명으로 사용할 수 없다!
