## 중첩 조건문

* nested conditional
* if 블록 안에 또 다른 if 블록이 있는 경우를 가리킨다.
* 들여쓰기로 논리를 구분한다.
* 가능하면 다른 형태로 고치는 것이 좋다.

```python
# ex1_nested conditional

x = 31
y = 21

if x == y :
    print("A")
else :
    if x < y : 
        print("B")
    else : 
        print("C")
```

```python
#correct_ex1
#if~elif~else문 사용

if x == y:
    print("A")
elif x < y :
    print("B")
else :
    print("C")
```

```python
#ex2_nested conditional

if x > 0 :
    if x < 15 :
        print("1")
    else :
        print("2")
else :
    print("3")
```

```python
#correct1_ex2
#해당 논리를 정리하면 0 < x < 15이므로 이를 나타내 준다.

if x > 0 and x < 15 : #논리연산자 사용
    print("1")
else :
    print("2")
```

```python
#correct2_ex2

if 0 < x < 15 : #파이썬에서만 사용 가능
    print("1")
else : 
    print("2")
```

### 바다코끼리 연산자 (Plus 요소, optional)
* walrus operator 
* 할당과 테스트를 동시에 하는 연산자이다.
* 코드에서는 :=와 같이 나타낸다.

```python
#ex_walrus

tweet_limit = 200
tweet_string = "Lullu" * 50
tweet_string
```

```python
diff = tweet_limit - len(tweet_string)
if diff > 0:
    print("Possible")
else: 
    print("Impossible")
```

```python
#위의 식을 바다코끼리 연산자를 사용해 다음과 같이 나타낼 수 있다.

if diff := tweet_limit - len(tweet_string) > 0:
        print("Possible")
```

## Slicing
* 지정 범위의 문자열을 추출하는 것이다.
* 문법은 다음과 같다 : 대괄호 + 오프셋[offset]
    - [start : end : step]
    - 오프셋은 항상 0부터 시작한다.   
    - 0은 생략한다.   
      ex) ~4까지 출력 : [:5]   
      ex) 처음부터 끝까지 2 스텝씩 : [::2]   
    - start는 포함하며, end는 포함하지 않는다. 즉, 크거나 같으며 작다.
    - 오프셋은 음수로도 가능하다(음수는 양수와 달리, -1부터 시작한다).   
      ex) h e l l o  
         -5 . . . -1   
          0 . . . 4
    - 오프셋은 거꾸로도 가능하다.   
      ex) 처음부터 끝까지 역순으로 : [::-1]

```python
sent = "Rose means love." #스텝
sent [:2]
```

```python
sent [2] #n번째 문자
```

```python
sent [::-1] #역순
```

```python
sent [:-5] 
```

```python
sent [::-5]
```

```python
len(sent) #length, 즉 길이를 알려준다! 이때, 띄어쓰기도 포함된다.
```
